## Import Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os, json
```

```python
df = pd.read_csv('superhero_info.csv')
df.head()
```

```python
df.info()
```

```python
dfp = pd.read_csv('superhero_powers.csv')
dfp.head()
```

```python
dfp.info()
```

## Seperate Hero and Publisher

```python
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head()
```

```python
df = df.drop(columns=['Hero|Publisher'])
df.head()
```

## Seperate Height and Weight, and make into int.

```python
df['Measurements'] = df['Measurements'].str.replace("'",'"')

df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
Height_Weight = df['Measurements'].apply(pd.Series)
Height_Weight
```

```python
df = pd.concat((df, Height_Weight), axis = 1)
df = df.drop(columns=['Measurements'])
df.head()
```

```python
df['Height'] = df['Height'].str.replace(' cm','')
df['Weight'] = df['Weight'].str.replace(' kg','')
```

```python
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
df.head()
```

```python
df.info()
```

```python
df = df.reindex(columns= ['Hero','Publisher','Gender','Race','Eye color','Hair color','Skin color','Height','Weight','Alignment'])
df.head()
# https://datagy.io/reorder-pandas-columns/
```

## One-hot-encoded Powers

```python
dfp.head()
```

```python
dfp['Powers_split'] = list(dfp['Powers'].str.split(','))
```

```python
exploded = dfp.explode('Powers_split')
exploded[['hero_names','Powers_split']].head()
```

```python
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    dfp[col] = dfp['Powers'].str.contains(col)
dfp.head()
```

```python
merged = pd.merge(df, dfp, left_on='Hero', right_on='hero_names', how='left')
merged.head()
```

## Questions

```python
# Compare the average weight of super powers who have Super Speed to those who do not.
Super = merged[merged['Super Speed']==True]['Weight'].mean()
NotSuper = merged[merged['Super Speed']==False]['Weight'].mean()
print(f'Super Speed Avg. Weight: {Super.round(3)}, Not So Super Speed Avg. Weight: {NotSuper.round(3)}')
```

```python
# What is the average height of heroes for each publisher?
merged.groupby('Publisher')['Height'].mean().round(3)
```
